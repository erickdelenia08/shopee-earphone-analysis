```python
# 5. Specifications Table
attr = []
for item in data:
    for v in item['attributes']:
        attr.append({
            'itemid':item['product']['itemid'],
            'attr_name':v['attr_name'],
            'attr_value':v['attr_value']
        })

df_attr = pd.DataFrame(attr)
```

```python
import pandas as pd
import numpy as np
import re

def warranty_to_days(val):
    if pd.isna(val):
        return 0
    val = str(val).lower().strip()

    if val in ['tidak ada garansi', 'tidak garansi', '.', '/', '']:
        return 0

    if 'hari' in val or 'x24 jam' in val:
        days = re.findall(r'\d+', val)
        return int(days[0]) if days else 1
    elif 'minggu' in val or 'week' in val or 'seminggu' in val:
        weeks = re.findall(r'\d+', val)
        return int(weeks[0])*7 if weeks else 7
    elif 'bulan' in val:
        months = re.findall(r'\d+', val)
        return int(months[0])*30 if months else 30
    elif 'tahun' in val:
        years = re.findall(r'\d+', val)
        return int(years[0])*365 if years else 365

    return 0  # default

# Terapkan langsung ke attr_value untuk masa garansi
mask = df_attr['attr_name'] == 'Masa Garansi'
df_attr.loc[mask, 'attr_value'] = df_attr.loc[mask, 'attr_value'].apply(warranty_to_days)
```

```python
mapping = {
    'tidak garansi': 'tidak ada garansi',
    'seller warranty': 'tidak ada garansi',
    'garansi supplier': 'garansi pabrikan',
    'garansi produsen': 'garansi pabrikan',
    'garansi resmi': 'garansi resmi',
    'garansi internasional': 'garansi resmi',
    'garansi toko': 'garansi penjual',
    'distributor':'garansi distributor',
    'garansi penjual': 'garansi penjual',
    'garansi seller': 'garansi penjual',
    'garansi distributor': 'garansi distributor',
    'garansi distributor  pt.harman indonesia': 'garansi distributor',
    'garansi resmi jete': 'garansi resmi',
}

# Terapkan mapping
mask = df_attr['attr_name'] == 'Jenis Garansi'
df_attr.loc[mask, 'attr_value'] = (
    df_attr.loc[mask, 'attr_value']
        .str.lower()
        .str.strip()
        .replace(mapping)
)
```
